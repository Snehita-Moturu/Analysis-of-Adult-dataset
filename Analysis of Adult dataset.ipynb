```
df1 = spark.read.format("csv").option("header", "true").load("dbfs:/FileStore/shared_uploads/sdiddi@gmu.edu/Adult_final.csv")
display(df1)
```

```
from functools import reduce

oldColumns = df1.schema.names
newColumns = [
     "age",
     "workclass",
     "finalweight",
     "education",
     "maritalstatus",
     "occupation",
     "relationship",
     "race",
     "sex",
     "capitalgain",
     "capitalloss",
     "hoursperweek",
     "nativecountry",
     "incomeclass"
 ]

df1 = reduce(lambda df1, idx: df1.withColumnRenamed(oldColumns[idx], newColumns[idx]), range(len(oldColumns)), df1)
df1.printSchema()
df1.select([count(when(isnan(c) | col(c).isNull(), c)).alias(c) for c in df1.columns]).show()
```

```
df1.display()
```

```
from pyspark.sql.functions import col

# Assuming your DataFrame is called "df"
df1 = df1.withColumn("hoursperweek", col("hoursperweek").cast("integer"))
df1 = df1.withColumn("age", col("age").cast("integer"))
df1 = df1.withColumn("finalweight", col("finalweight").cast("integer"))
df1 = df1.withColumn("capitalgain", col("capitalgain").cast("integer"))
df1 = df1.withColumn("capitalloss", col("capitalloss").cast("integer"))
```

```
df1.display()
```

```
df1.printSchema()
```

```
#Encoding the data for GENDER where 0 = Female, 1 = Male
from pyspark.sql.functions import when
 
df1 = df1.withColumn("incomeclass", when(df1["incomeclass"] == "<=50K", 0)
                                            .otherwise(1))
 
df1.select("incomeclass").distinct().show()
```

```
df1.display()
```

```
from pyspark.sql.functions import col

new_df = df1.filter(col('workclass') == 'Private')
print(new_df.filter(col('incomeclass') == 1).count())
new_df.filter(col('incomeclass') == 1).show(2)
```

```
#Checking the ratio of male and female according to their income
import pyspark.sql.functions as F

class_sex = df1.crosstab('incomeclass', 'sex')
class_sex.display()
```
